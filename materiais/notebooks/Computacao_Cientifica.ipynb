# Principais módulos e métodos

## Norma de um vetor

A **norma** de um vetor $\vec{AB}$, ou módulo, é o _comprimento_ do segmento $AB$. A norma de um vetor é denotada da seguinte forma:

$$||\vec{AB}||$$

Se $||\vec{AB}||$ = 1, dizemos que o vetor $\vec{AB}$ é _unitário_. Vamos a um exemplo:

Dado o vetor $\vec{u} = (3, 3, 1)$, calcule $\|\vec{u}\|$.

Primeiro, devemos saber que $\|\vec{u}\|$ = $\sqrt{\langle \vec{u}, \vec{u}\rangle}$. A notação $\langle \vec{u}, \vec{u}\rangle$ denota o produto interno do vetor $\vec{u}$ por ele mesmo. 

Calculando o produto interno $\langle \vec{u}, \vec{u}\rangle$, temos que:

$$\langle \vec{u}, \vec{u}\rangle  = \langle (3, 3, 1), (3, 3, 1)\rangle  = (3\cdot3) + (3\cdot3) + (1\cdot1) = 19$$

Logo, 

$$||\vec{u}|| = \sqrt{\langle \vec{u}, \vec{u}\rangle } = \sqrt{19} \approx 4.3588$$

Usaremos o mesmo vetor anterior para calcular a norma, através do algoritmo abaixo.

```
import numpy as np
from scipy.linalg import norm # Importando a função 'norm'

vector = np.array([3, 3, 1]) # Declarando o vetor
print('Vetor: \n', vector)
print('\nNorma do vetor: \n', round(norm(vector), 4)) # Resultado da norma com 4 casas decimais
```

## Determinante de uma matriz

O **determinante** de uma matriz é um número real que está associado a uma matriz **quadrada** e esse valor é único, isto é, para cada matriz quadrada de ordem $n$ temos um único número real que se associa a ela. Vamos ao seguinte exemplo:

Considere a matriz de ordem 2 

$$A = \begin{pmatrix} 
  1 & 5 \\ 
  -2 & 5
  \end{pmatrix}$$

Vamos calcular o determinante

$$ det(A) = \left[ \begin{array}{ccc}
 1 & 5 \\ 
 -2 & 5 
\end{array} \right]$$

Primeiramente, calculamos o produto da sua diagonal principal, isto é, 

$$1 * 5 = 5$$

Logo após, calculamos o produto da sua diagonal secundária, que é

$$-2 * 5 = -10$$

Por fim, temos que o determinante é dado pela fórmula

$$det(A) = \mathrm{diagonal \ principal} - \mathrm{diagonal \ secundária}$$

Logo, 

$$det(A) = 5 - (-10) = 15$$

Vamos usar a mesma matriz no código abaixo.

```
import numpy as np
from scipy.linalg import det # Importando a função 'det'

matrix = np.array([[1, 5], [-2, 5]]) # Declarando a matriz
print('Matriz: \n', matrix)
print('\nDeterminante: \n', det(matrix))
```

## Inversa de uma matriz

A matriz inversa tem a propriedade de quando multiplicada a matriz base resultar na matriz identidade.

$$\begin{equation}
  A^{-1}\ ×\ A = I
\end{equation}$$

Ela é definida como:

$$A^{-1} = \frac{1}{det(A)} \cdot adj(A)$$

Vamos a um exemplo de código.

```
import numpy as np
from scipy.linalg import inv # Importando função 'inv'

matrix = np.array([[3, -1, 0], [0, 2, 1], [1, 1, 0]]) # Declarando a matriz
print('Matriz inversa: \n', inv(matrix)) # Calculando a inversa
print('\nMatriz identidade: \n', np.dot(matrix, inv(matrix))) # Confirmando o cálculo
```

## Funções de Otimização

Para alguns calculos utilizados para algoritmos, sempre temos o intuito de otimizar o processo de tal forma que possamos reduzir custo, tempo e linha de código, para isso o **SCIPY** traz um modulo justamente com funções que nos ajudam no processo de otimização, utilizando a ideia de **MINIMIZAÇÃO** e **MAXIMIZAÇÃO** de funções, muito utilizado em processos de aprendizagem de máquina.

```
from scipy.optimize import minimize

f = lambda x: x ** 2
x0 = 10

res = minimize(f, x0)
print(res.x)
```

```
x0 = 1
res = minimize(f, x0)
print(res.x)
```

```
x0 = 10
bounds = [(0, None)]

res = minimize(f, x0, bounds=bounds)
print(res.x)
```

Vale ressaltar que, para problemas onde **desconhecemos** parâmetros de uma função, como o da regressão linear, necessitamos estima-los, e para isso a função minimize pode nos ajudar também a achar os coeficientes dessa reta, da seguinte forma:

```
import numpy as np

np.random.seed(42)
X = np.random.random(1000)
y = 3 * X + 2 + np.random.normal(size=1000)

betas = np.random.normal(size=2)

sq_err = lambda b, x, y: np.mean(((x*b[1] + b[0]) - y) ** 2)

res = minimize(sq_err, betas, args=(X, y))
print(res.x)
```

## Distribuições de probabilidade

Distribuições de probabilidade são funções, que para o caso contínuo (definido nos reais), temos a função densidade de probabilidade, com ela podemos definir probabilidades de um intervalo de observações aconteça, ou o valor esperado dado aquele conjunto de observações, caso sigam uma distribuição de probabilidade. Vamos utilizar a distribuição **NORMAL** como exemplo. Por notação, dizemos que uma **Variável aleatória X**  segue uma distribuição normal, com média $\mu$ e variância $\sigma^{2}$, da seguinte forma:

$$
X \sim N(\mu, \sigma^{2}) 
$$

Com **Função de densidade de probabilidade (fdp)** definido da seguinte forma

$$
f(x | \mu, \sigma^{2}) = \frac{1}{\sqrt{2 \pi \sigma^{2}}} \cdot e^{- \frac{(x - \mu)^{2}}{2 \sigma^{2}}}
$$

Onde cada valor assumido por **X** é uma observação do seu *set* de dados.

```
# Importando o método da normal
from scipy.stats import norm
```

```
# Calculando alguns valores importantes
mean, var, skew, kurt = norm.stats(moments='mvsk')
print('média: {}, variância: {}, assimetria: {}, curtose: {}'.format(mean, var, skew, kurt))
```

A normal geranda acima, é uma normal onde a média é igual a 0 ($\mu = 0$) e variância igual a 1 ($\sigma^{2} = 1$). Chamamos essa normal de **Normal Padrão**, por notação temos:

$$
X \sim N(0,1)
$$

Com **Função densidade de probabilidade** definido da seguinte forma:

$$
f(x) = \frac{1}{\sqrt{2\pi}} \cdot e^{-x^{2}}
$$

Se quisermos mudar os parâmetros dessa normal, de escala ($\sigma$) e de localização ($\mu$), da seguinte forma:

```
mean, var = norm.stats(loc=2, scale=0.5, moments='mv')
print('média: {}, variância: {}'.format(mean, var, skew, kurt))
```

Vamos calcular a densidade para um ponto $x = 4$, em uma normal de parâmetros:

$$
\sigma = 0.5 \text{ e } \mu = 5 
$$

```
x = 4
print(norm.pdf(x, loc=5, scale=0.5))
```

Vamos plotar o gráfico dessa densidade, onde cada $f(x)$ representa a probabilidade de cada elemento $x$. Então:

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4, 1000)
plt.plot(x, norm.pdf(x, loc=2, scale=0.5))
plt.show()
```

Além disso, temos um resultado muito importante que é a **Função de probabilidade acumulada (cdf)**, como o nome ja diz, é uma função que define a probabilidade acumulada da **pdf**. Então:

```
plt.plot(x, norm.cdf(x, loc=2, scale=0.5))
plt.show()
```

Além disso tudo visto, temos  algo bem interessante, que é a possibilidade da  construção de interavalos de confiança para a verdadeira média das nossas observações, da seguinte forma:}

```
print(norm.interval(0.95, loc=5, scale=0.5))
```

Por fim, mas não falando sobre tudo ainda que temos na biblioteca, podemos gerar valores a partir de um distribuição **Normal** com a média que quisermos e vâriancia escolhida. Então:

```
rv = norm(loc=2, scale=0.5)

samples = rv.rvs(size=1000)
plt.hist(samples)
plt.show()

print(samples[:100])
```

## Testes de Hipóteses

Testes de hipóteses são muito importantes para verificações de suposições, como verificar se a **média** que observamos é a **verdadeira média** dos nossos dados, dado que muitas vezes temos so amostras e não a população inteira, o teste para esse tipo de caso ficaria da seguinte forma :

$$
H_{0}: \bar{x}  = \mu \\
H_{1}: \bar{x}  \neq \mu
$$

![img](https://lh3.googleusercontent.com/proxy/R83IIBoRXJXCR0OIm1zp3RRKctjqivE8kFQlfUk279_FEMY_vx7At9siHaa2S26m_59C5d8OCZ6ogMF5Ze9h1EIlp1UvuHmhJLViAY4sADYakIqUc9CHKujYLwEFYa65zFYEFtDs3DtvW8GYaQhYtxV2-Mzf7g-YHOKR5bgyEIfzGLTozurCm-jsOIYBpSpNGbGnVOHfKO8mmx3SaUJZo4x19j432gHyj9qNc2-nWkgxT2eDNxpmHoJD)


Sendo 
$$
H_{0}: \text{ é a nossa hipótese que queremos testar}\\
H_{1}: \text{ é a nossa hipótese alternativa} \\
\alpha: \text{ é o nosso nivel de significancia para realizarmos o teste} \\
Z_{\frac{\alpha}{2}}: \text{ é o valor calculado da nossa estatística do teste}
$$

Esse é um exemplo de teste bilateral que usamos para verificar se nossa média observada é o verdadeiro valor da média populacional. Para esse teste, consideramos o teste **t-student**, da seguinte forma:

```
from scipy.stats import ttest_1samp

np.random.seed(42)

x = norm.rvs(loc=5, scale=10, size=1000) # Gerando uma normal com loc = 5, e scale = 10
print(
    'Não rejeita a hipótese nula: {}'.format(ttest_1samp(x, 5))
)
print()
print('Esse teste testamos a hipótese de que a média amostral é igual 5, a partir ')
print('de um conjunto de dados que esta sendo gerado a partir de uma normal de ')
print('parâmetros N(5,(10)^{2}), logo intuitivamente, a nossa hipótese nula não é ')
print('rejeitada.')
print()
print()
print(
    'Rejeita a hipótese nula: {}'.format(ttest_1samp(x, 0))
)
print()
print('Esse teste testamos a hipótese de que a média amostral é igual 0, a partir ')
print('de um conjunto de dados que esta sendo gerado a partir de uma normal de ')
print('parâmetros N(5,(10)^{2}), logo intuitivamente, a nossa hipótese nula foi ')
print('rejeitada.')
print()
print()
```

Além do teste **t-student**, o modulo apresenta inúmeros outros testes estatísticos muito importantes, como o de **Wilcoxon**, testes de normalidades, **Shapiro** ou **Qui-Quadrado**, entre outros.

## Outras distribuições de probabilidade

### Gamma
A função de densidade de probabilidade de Gamma é definida por:
$$
f(x, a) = 
\frac{x^{a-1} \exp(-x)}{\Gamma(a)}
$$

sendo $x \ge 0$ e $a \gt 0$. Aqui, a função Gamma é referida por $\Gamma(a)$ e toma $a$ como parâmetro de forma.

```
from scipy.stats import gamma
import matplotlib.pyplot as plt
import numpy as np
```

```
a = 1.99    #Definindo um valor para "a"

mean, var, skew, kurt = gamma.stats(a, moments='mvsk')    #Calculando os primeiros momentos
print('média: {}, variância: {},  assimetria: {}, curtose: {}'.format(mean, var, skew, kurt))
```

```
x = np.linspace(gamma.ppf(0.01, a), gamma.ppf(0.99, a), num = 100)    #Dispondo a função de densidade de probabilidade (pdf em inglês) "pdf gamma"
rv = gamma(a)    #Dispondo a distribuição com seu valor inicial. Chamaremos de "pdf congelado" 
r = gamma.rvs(a, size=1000)    #Gerando valores aleatórios para "r"
```

```
fig, ax = plt.subplots(1, 1)    #Plotando o gráfico com locações (parâmetro loc) diferentes

ax.plot(x, gamma.pdf(x, a, loc = 0), 'r-', lw=5, alpha=0.6, label='pdf gamma loc = 0')  #O padrão é a locação ser 0!
ax.plot(x, gamma.pdf(x, a, loc = 1), 'b-', lw=5, alpha=0.6, label='pdf gamma loc = 1')
ax.plot(x, gamma.pdf(x, a, loc = 2), 'g-', lw=5, alpha=0.6, label='pdf gamma loc = 2')
ax.plot(x, gamma.pdf(x, a, loc = 3), 'y-', lw=5, alpha=0.6, label='pdf gamma loc = 3')
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='pdf congelado')
ax.legend(loc = 'best', frameon = False)
ax.hist(r, density = True, histtype = 'stepfilled', alpha = 0.2)
plt.show()
```

```
fig, ax = plt.subplots(1, 1)    #Plotando o gráfico com escalas (parâmetro scale) diferentes

ax.plot(x, gamma.pdf(x, a, scale = 1), 'r-', lw=5, alpha=0.6, label='pdf gamma scale = 1') #O padrão é a escala ser 1!
ax.plot(x, gamma.pdf(x, a, scale = 2), 'b-', lw=5, alpha=0.6, label='pdf gamma scale = 2')
ax.plot(x, gamma.pdf(x, a, scale = 3), 'g-', lw=5, alpha=0.6, label='pdf gamma scale = 3')
ax.plot(x, gamma.pdf(x, a, scale = 4), 'y-', lw=5, alpha=0.6, label='pdf gamma scale = 4')
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='pdf congelado')
ax.legend(loc = 'best', frameon = False)
ax.hist(r, density = True, histtype = 'stepfilled', alpha = 0.2)
plt.show()
```

### Lognormal

$$

\frac{1}{s x \sqrt{2\pi}}
\exp\left(-\frac{\log^2(x)}{2s^2}\right)
$$
para $x \gt 0$ e $s \gt 0$.

```
from scipy.stats import lognorm
import matplotlib.pyplot as plt
import numpy as np
```

```
fig, ax = plt.subplots(1, 1)    #Plotando o gráfico com locações (parâmetro loc) diferentes

ax.plot(x, lognorm.pdf(x, s, loc = 0), 'r-', lw=5, alpha=0.6, label='pdf lognorm loc = 0')  #O padrão é a locação ser 0!
ax.plot(x, lognorm.pdf(x, s, loc = 1), 'b-', lw=5, alpha=0.6, label='pdf lognorm loc = 1')
ax.plot(x, lognorm.pdf(x, s, loc = 2), 'g-', lw=5, alpha=0.6, label='pdf lognorm loc = 2')
ax.plot(x, lognorm.pdf(x, s, loc = 3), 'y-', lw=5, alpha=0.6, label='pdf lognorm loc = 3')
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='pdf congelado')
ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```

## Normal
A função de densidade de probabilidade da Normal é definida por:
$$
f(x) = \frac{\exp(-x^2/2)}{\sqrt{2\pi}}
$$
para um número $x$ real.

```
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```
mean, var, skew, kurt = norm.stats(moments='mvsk')    #Calculando os primeiros momentos
print('média: {}, variância: {},  assimetria: {}, curtose: {}'.format(mean, var, skew, kurt))
```

```
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), num  = 100)    #Dispondo a função de densidade de probabilidade "pdf norm"
rv = norm()    #Dispondo a distribuição com seu valor inicial. Chamaremos de "pdf congelado"
r = norm.rvs(size=1000)    #Gerando valores aleatórios para "r"
```

```
fig, ax = plt.subplots(1, 1)    #Plotando o gráfico com locações (parâmetro loc) diferentes

ax.plot(x, norm.pdf(x, loc = 0), 'r-', lw=5, alpha=0.6, label='pdf norm loc = 0')  #O padrão é a locação ser 0!
ax.plot(x, norm.pdf(x, loc = 1), 'b-', lw=5, alpha=0.6, label='pdf norm loc = 1')
ax.plot(x, norm.pdf(x, loc = 2), 'g-', lw=5, alpha=0.6, label='pdf norm loc = 2')
ax.plot(x, norm.pdf(x, loc = 3), 'y-', lw=5, alpha=0.6, label='pdf norm loc = 3')
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='pdf congelado')
ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```
