# **Introdução ao Matplotlib**

Matplotlib é uma biblioteca do Python utilizada para visualizar dados. Ela oferece uma grande variedade de opções, desde histogramas e gráficos de barras até a visualização de superfícies, de maneira simples e rápida através de funções. 

A documentação completa do Matplotlib pode ser encontrada [aqui](https://matplotlib.org/index.html#).

## **Importar a biblioteca**

```
#Importar o Matplotlib e o Numpy

import matplotlib.pyplot as plt

import numpy as np
```

## **Pyplot**

A maneira mais simples de visualizar os dados é através de funções disponíveis no módulo Pyplot. Com o .plot, por exemplo, é possível gerar rapidamente um gráfico de linha de uma função exponencial.

```
#Plot de um gráfico

x = np.arange(0,10,1)            #10 valores para o eixo x
y = x**2                         #Valores do eixo y

print("Eixo x: ",x)
print("Eixo y: ",y)

plt.plot(x, y)
plt.show()
```

Os elementos do gráfico podem ser facilmente editados através de outras funções, como o plt.title()

```
x = np.arange(0,10,0.1)
y = np.sin(x)

plt.plot(x, y)

plt.title('Função seno')             #Adiciona título
plt.ylabel("sen(x)")                 #Adiciona nome ao eixo y
plt.xlabel("x")                      #Adiciona nome ao eixo x
plt.axis([-1, 11, -1.2, 1.2])        #Altera os limites dos eixos (xmin, xmax, ymin, ymax)
plt.grid(True)                       #Adiciona grid

plt.show()
```

Também é possível fazer múltiplos plots ao mesmo tempo e alterar a escala do gráfico de linear para log, por exemplo.

```
#Código retirado da documentação do Matplotlib

# cria valores entre 0 e 1
y = np.random.normal(loc=0.5, scale=0.4, size=1000)     #distribuição normal com centro em 0,5 e desvio padrão = 0,4
y = y[(y > 0) & (y < 1)]                                #seleciona os valores entre 0 e 1
y.sort()                                                #coloca-os em ordem crescente
x = np.arange(len(y))                                   #eixo x

plt.figure(figsize = (10, 5))                             #cria a figura onde estarão os gráficos, onde pode-se alterar o tamanho

# linear
plt.subplot(1,2,1)                                      #indica as dimensoes do grid e a posição do gráfico (1 linha, 2 colunas, primeiro gráfico)
plt.plot(x, y)
plt.yscale('linear')                                    #define a escala do eixo y para linear, que é a padrão 
plt.title('linear')
plt.grid(True)


# log
plt.subplot(1,2,2)
plt.plot(x, y)
plt.yscale('log')                                       #define a escala do eixo y para log
plt.title('log')
plt.grid(True)


plt.show()
```

Outros tipos de gráficos estão disponíveis no pacote além do gráfico de linha:

```
#Gráfico de barra, muito utilizado para visualização de dados categóricos

x = ['A','B', 'C']
y = [10, 20, 30]

plt.bar(x,y)
plt.title('Gráfico de barras')
plt.xlabel('Categorias')
plt.ylabel('Valores')

plt.show()
```

```
#Gráficos de dispersão, utilizado para visualizar relações entre variáveis

from sklearn.datasets import make_blobs

x, y = make_blobs(n_features=2, centers=3,                #Cria um dataset com 3 clusters
                  random_state = 42)                

plt.scatter(x[:,0], x[:,1], c = y)                        #Gráfico de dispersão entre as duas variáveis do dataset, destacando-se o label através de cores
plt.title('Dispersão')
plt.xlabel("x1")
plt.ylabel('x2')

plt.show()
```

```
#Comparação entre duas distribuições utilizando histogramas

x1 = np.random.randn(1000)                                                      #Distribuição normal
x2 = np.random.beta(3, 3, size = 1000)                                          #Distribuição Beta

plt.figure(figsize = (7,7))

plt.hist(x1, bins = 20, alpha = 0.5, label = "Dist. Normal")
plt.hist(x2, bins = 10, color = 'red', alpha = 0.3, label = "Dist. Beta")
plt.title('Histograma')
plt.legend()

plt.show()
```

Esses são alguns dos exemplos de gráficos e personalizações mais simples. O Pyplot disponibiliza uma infinidade de opções de forma simples, que podem ser encontradas na documentação da biblioteca.

## **Interface Orientada a Objetos**

Caso seja necessário um maior controle dos gráficos a serem gerados, é possível utilizar uma interface orientada a objetos, que podem envolver ou não o uso de funções do Pyplot. Nesse nível de hierarquia, encara-se a figura e seus elementos como objetos, que podem ser criados e modificados através de código.

Os passos principais são a criação da figura (fig), onde os elementos estarão inseridos, e do objeto da classe *axes*, que pode ser pensado como o gráfico de fato, o lugar onde os dados serão representados. É através de métodos dessa classe que se adiciona e se modifica boa parte dos elementos do gráfico.

Dentro do *axes*, pode-se acessar os objetos *axis*, que são os eixos x, y e z do gráfico, e alterá-los.

```
plt.subplots()
```

```
fig, ax = plt.subplots()                                                            #Criação da figura e do axes

x = np.arange(0, 10, 0.5)
y_lin = x
y_sqr = x**2


ax.plot(x, y_lin, marker = 'o', c = 'green', ls = 'dashed', label = 'Linear')        #Plota os dados no ax
ax.plot(x, y_sqr, 'rs', label = 'Quadrática')                                        # 'rs' = red squares


ax.set_title('Exponencial x Linear')                                                #Modifica seus elementos gerais
ax.set_xlabel('x')
ax.set_ylabel('y')


ax.legend(loc = 'upper left')                                                       #Adiciona legenda e numa posição específica


ax.xaxis.set_ticks_position('top')                                                  #Edita os eixos
ax.yaxis.set_ticks_position('right')


plt.show()
```

## **Visualização em 3D**

O Matplotlib conta com uma bilbioteca para viabilizar a criação de gráficos 3D, chamado *mplot_3d*. A partir dela, pode-se passar o argumento projection = "3d" no momento da criação do axes.

```
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = plt.axes(projection="3d")

plt.show()
```

Os métodos utilizados para plotar em 3D são bem parecidos com os plots em 2D. Por exemplo, em vez de usar o método scatter, usa-se o método scatter3D. 

```
fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection = '3d')

z_linha = np.linspace(0, 15, 1000)
x_linha = np.cos(z_linha)
y_linha = np.sin(z_linha)
ax.plot3D(x_linha, y_linha, z_linha)

z_pontos = 15 * np.random.random(100)
x_pontos = np.cos(z_pontos) + 0.1 * np.random.randn(100)
y_pontos = np.sin(z_pontos) + 0.1 * np.random.randn(100)
ax.scatter3D(x_pontos, y_pontos, z_pontos, c=z_pontos, cmap='hsv')

plt.show()
```

Gráficos 3D também são muito utilizados para visualizar superfícies.

```
from matplotlib import cm                                         #Importa o mapa de cores

fig = plt.figure(figsize = (15,10))
ax = plt.axes(projection = '3d')

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
